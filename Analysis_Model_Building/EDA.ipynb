```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
from tqdm import tqdm
import json
```

```python
#@title Data Loading
train = pd.read_csv('train.csv')
```

```python
cols1 = [f"CONC{i}" for i in range(5)]
cols2 = ['CONC6', 'CONC8', 'CONC10', 'CONC12', 'CONC14']

condition = train['SAMPLE_DATA_TYPE'] == 'cell_red'

temp = train.loc[condition, cols1].copy()

train.loc[condition, cols2] = temp.to_numpy() # just df assigment to df.loc is not wortking
train.loc[condition, cols1] = None
```

```python
#@title CONC error per sample for only red autofluorescence
conc = train[[f"CONC{i}" for i in range(15)]]
conc.describe().T[['mean',	'std']]
```

```python
#@title CONC error per sample for only red autofluorescence
conc = train[train['SAMPLE_DATA_TYPE'] == 'cell_red'][[f"CONC{i}" for i in range(15)]]
conc.describe().T[['mean',	'std']]
```

```python
#@title CONC error per sample for only tox21-err-p1 assay
conc = train[train['SAMPLE_DATA_TYPE'] != 'cell_red'][[f"CONC{i}" for i in range(15)]]
conc.describe().T[['mean',	'std']]
```
